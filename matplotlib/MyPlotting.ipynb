```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12,4))
fig, bx1 = plt.subplots(1, 1, constrained_layout=True, figsize=(24,8))
ax1.set_xlabel("width"); ax1.set_ylabel("height")
ax1.plot(x, y)
ax2.plot(x + 5, y + 1)
bx1.plot(x, y)
```
