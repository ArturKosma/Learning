```python
import matplotlib.animation
import matplotlib.pyplot as plt
from itertools import count
import random
import time

plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['figure.dpi'] = 150  
plt

fig, ax = plt.subplots()
x_value = []
y_value = []
index = count();
def animate(t):
    x_value.append(next(index))
    y_value.append(random.randint(0,10))
    ax.cla()
    ax.plot(x_value,y_value)
    ax.set_xlim(0,10)

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=10, interval = 50)
plt.close()

ani
```

```python
fig, ax = plt.subplots()

last_time = [time.time()]
start_time = time.time()
position = [0, 0]

def animate_dots(t):

    current_time = time.time()
    delta_time = current_time - last_time[0]
    last_time[0] = current_time
    elapsed_time = current_time - start_time
    
    position[0] += delta_time
    position[1] += delta_time 
    ax.cla()
    ax.scatter(position[0], position[1], c='r')
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    
    ax.text(9.8, 9.5, f"FPS: {(delta_time * 1000):.0f}", fontsize=8, verticalalignment='center', horizontalalignment='right')
    ax.text(9.8, 9.2, f"Time Elapsed: {elapsed_time:.2f}s", fontsize=8, verticalalignment='center', horizontalalignment='right')
    
    return ax

fig.set_size_inches(8, 6)
anim = matplotlib.animation.FuncAnimation(fig, animate_dots, frames=range(60), interval=16.67)
plt.close()

anim
```
